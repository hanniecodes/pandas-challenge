```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
# school_data_complete2=school_data_complete
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#calclulate the total number of schools 
total_schools = school_data_complete["school_name"].nunique()
#calculate the total number of students 
total_students = school_data_complete["Student ID"].nunique()
#calculate the total budget 
total_budget = school_data["budget"].sum()
# Calculate the average math score
average_math = school_data_complete["math_score"].mean()
# Calculate the average reading score
average_reading = school_data_complete["reading_score"].mean()
# Calculate the percentage of students with a passing math score (70 or greater)
passingmath = len(school_data_complete [school_data_complete["math_score"]>=70])
per_passingmath= (passingmath/total_students)*100
# Calculate the percentage of students with a passing reading score (70 or greater)
passingread = len(school_data_complete [school_data_complete["reading_score"]>=70])
per_passingread= (passingread/total_students)*100
# Calculate the percentage of students who passed math and reading (% Overall Passing)
#check with tutor 
#This is wrong
# passmathread=(passingmath & passingread)/total_students
passmathread=(student_data[(student_data['math_score'] >= 70) & (student_data['reading_score'] >= 70)]['Student ID'].count()/total_students)*100
```

```python
# Create a dataframe to hold the above results
schoolsummary_df = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math],
    "Average Reading Score": [average_reading],
    "% Passing Math": [per_passingmath],
    "% Passing Reading": [per_passingread],
    "% Overall Passing": [passmathread]})
```

```python
# Optional: give the displayed data cleaner formatting
schoolsummary_df["Total Budget"] = schoolsummary_df["Total Budget"].map("${:,.2f}".format)

schoolsummary_df[["Total Schools",
      "Total Students",
      "Total Budget",
      "Average Math Score",
      "Average Reading Score",
      "% Passing Math",
      "% Passing Reading",
      "% Overall Passing"]]
```

```python
# Create an overview table that summarizes key metrics about each school, including:
# School Name
by_school_df = school_data_complete.groupby(["school_name"])
# School Type
school_type=by_school_df["type"].first()
# Total Students
student_total=by_school_df["Student ID"].count()
# Total School Budget
school_budget=by_school_df["budget"].first()
# Per Student Budget
perstudentbudget=school_budget/student_total
perstudentbudget2=school_budget/student_total
```

```python
# Average Math Score
averagemath=by_school_df["math_score"].mean()
# Average Reading Score
averageread=by_school_df["reading_score"].mean()
# % Passing Math
passingmath = school_data_complete [school_data_complete["math_score"]>=70].groupby(["school_name"])["math_score"].count()
per_passingmath= (passingmath/student_total)*100
# % Passing Reading
passingreading = school_data_complete [school_data_complete["reading_score"]>=70].groupby(["school_name"])["reading_score"].count()
per_passingread= (passingreading/student_total)*100
# % Overall Passing (The percentage of students that passed math and reading.)
percent_overall = (school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)].groupby("school_name")['Student ID'].count()/student_total)*100 
# percent_overall=(student_data[(student_data['math_score'] >= 70) & (student_data['reading_score'] >= 70)].groupby("school_name")['Student ID'].count()/student_total)*100                   
```

```python
# Create a dataframe to hold the above results
by_school_df = pd.DataFrame({
    "School Type": school_type,
    "Total Students": student_total,
    "Total School Budget": school_budget,
    "Per Student Budget": perstudentbudget,
    #creating this here for the 3rd to last exercise
    "Per Student Budget 2":perstudentbudget2,
    "Average Math Score": averagemath,
    "Average Reading Score": averageread,
    "% Passing Math": per_passingmath,
    "% Passing Reading": per_passingread,
    "% Overall Passing": percent_overall})
```

```python
by_school_df["Total School Budget"] = by_school_df["Total School Budget"].map("${:,.2f}".format)
by_school_df["Per Student Budget"] = by_school_df["Per Student Budget"].map("${:,.2f}".format)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

# Create a pandas series for each grade. Hint: use a conditional statement.

# Group each series by school
grade9 = school_data_complete.loc[school_data_complete["grade"] == "9th"].groupby(["school_name"])["math_score"].mean()
grade10 = school_data_complete.loc[school_data_complete["grade"] == "10th"].groupby(["school_name"])["math_score"].mean()
grade11 = school_data_complete.loc[school_data_complete["grade"] == "11th"].groupby(["school_name"])["math_score"].mean()
grade12 = school_data_complete.loc[school_data_complete["grade"] == "12th"].groupby(["school_name"])["math_score"].mean()
# Combine the series into a dataframe


bygrades_df=pd.DataFrame({"Ninth":grade9, 
                         "Tenth":grade10, 
                         "Eleventh":grade11, 
                         "Twelth":grade12})

bygrades_df["Ninth"] = bygrades_df["Ninth"].map("{:,.2f}".format)
bygrades_df["Tenth"] = bygrades_df["Tenth"].map("{:,.2f}".format)                                            
bygrades_df["Eleventh"] = bygrades_df["Eleventh"].map("{:,.2f}".format)                                               
bygrades_df["Twelth"] = bygrades_df["Twelth"].map("{:,.2f}".format)  

bygrades_df.head(15)

# Optional: give the displayed data cleaner formatting
```
